```mysql
-- 将goods表中的cate_name添加到cate_name表的name
insert into good_catname (name) select cate_name from goods group by cate_name;
-- 将goods表中的cate_name替换为cate_name表中的id
update goods as g inner join cate_name as c on g.cate_name=c.name set g.cate_name = c.id;

-- 插入外键：
-- 将字段数据类型修改为相同的
alter table goods change cate_name cate_id int unsigned not null;
-- 将goods表的cate_id设为外键并于goods_cates表的id相关联
alter table goods add foreign key (cate_id) references goods_cates(id)

-- 取消外键：
-- 查看外键名称：
show create table goods
-- 通过外键名称取消外键：
alter table goods drop foreign key 外键名称
```
