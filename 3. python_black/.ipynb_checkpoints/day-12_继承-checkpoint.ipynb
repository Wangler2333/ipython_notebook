### __init__
- 初始化方法
- 调用时自动执行

```python
class LiDuo:
    def __init__(self, new_height):
        self.height = new_height
    def height_print(self):
        print("LiDuo's height is %.2fm" % self.height)

if __name__ == "__main__":
    liduo = LiDuo(1.65)
    liduo.height_print()
```

# 继承（重点）


```python
# 无继承
class Animal:
    def eat(self):
        print("吃")
    def drink(self):
        print("喝")
    def run(self):
        print("跑")
    def sleep(self):
        print("睡")

class Dog:
    def eat(self):
        print("吃")
    def drink(self):
        print("喝")
    def run(self):
        print("跑")
    def sleep(self):
        print("睡")
    
    def dark(self):
        print("汪汪")

wangcai = Animal()
wangcai.eat()

wangcai2 = Dog()
wangcai2.dark()
```

```python
# 有继承
class Animal:
    def eat(self):
        print("吃")
    def drink(self):
        print("喝")
    def run(self):
        print("跑")
    def sleep(self):
        print("睡")
        
class Dog(Animal):
    def dark(self):
        print("jiaojaio")
        
hahah = Dog()
hahah.run()
```
