## 继承

```python
class A :
    def ap(self):
        print("A")
    
class B(A):
    def bp(self):
        print("B")
    
class C(B):
    def cp(self):
        print("c")
    
c = C()
c.ap()
c.bp()
c.cp()
```

### 父类重写及扩展

```python
# super(父类中传入的参数).bp(父类方法中传入的参数)
# super()代指父类 super().bp()相当于调用父类方法
class A :
    def ap(self):
        print("A")
    
class B(A):
    def bp(self):
        print("B")
    
class C(B):
    def cp(self):
        print("C")
    def bp(self):
        print("bp")
        print("super调用开始")
        super().bp()
        super().ap()
        print("super调用完成")
c = C()
c.bp()
```

### 隐藏属性和方法的子类调用

```python
class A :
    def ap(self):
        print("A")
    
class B(A):
    def __bp(self):
        print("B")
    
class C(B):
    def cp(self):
        print("c")
        
c = C()
c._B__bp()
```

## 多继承

```python
class A :
    def ap(self):
        print("A")
    
class B :
    def bp(self):
        print("B")
    
class C(A, B):
    pass

c = C()
c.ap()
c.bp()
```

```python
class A :
    def ap(self):
        print("A")
    
class B :
    def ap(self):   # 与A类中相同
        print("B")
    
class C(A, B): # 继承顺序决定调用时采用的方法
    pass

c = C()
c.ap()
print(C.__mro__) # 查看继承执行顺序
```

# 多态
- 封装
- 继承
- 多态：不同的子类对象调用相同的父类方法，产生不同的执行结果（继承+重写），即调用技巧


```python
class Dog(object):
    def __init__(self, name):
        self.name = name
    def game(self):
        print("%s在地上玩耍"% self.name)
        
class XiaoTianDog(Dog):
    def game(self):
        print("%s飞到天上..玩耍"% self.name)
        
class Person(object):
    def __init__(self, name):
        self.name = name
    def game_with_dog(self, dog):
        print("%s和%s一起玩，"%(self.name,dog.name),end="")
        dog.game()
        
xiaoming = Person("小明")
wangwang = Dog("旺财")
feiwangcai = XiaoTianDog("飞天旺财")

xiaoming.game_with_dog(wangwang)  # 多态用法
xiaoming.game_with_dog(feiwangcai)
```

### 类方法
- @classmethod
- def 方法名(cls)：
- ...

**cls.属性/cls.方法，可以访问类属性/方法**

```python
class B(object):
    
    @classmethod
    def bp(cls):
        print("这是个类方法")
        
B.bp()
```

```python
class BB(object):
    aa = "这是个类属性"
    @classmethod
    def cc(cls):
        print("这是类方法，调用了'%s'" % cls.aa)

print(BB.aa)
BB.cc()
```

# 单例
- 单例设计模式：单个实例，让类创建的对象，在系统中只有唯一的一个实例
- __new__内置静态方法：1.为对象分配空间；2.返回对象的引用

```python
# __new__方法重写
class MusicPlayer(object):
    
    def __new__(cls,*args,**kwargs):
        # 创建时调用
        print("创建对象，分配空间")
        return super().__new__(cls)
    def __init__(self):
        print("播放器初始化")
        
mp = MusicPlayer()
```

```python
# 正常类的创建
class MP(object):
    pass

mp1 = MP()
mp2 = MP()
print(mp1)
print(mp2)# 地址不同
```

```python
# 实现单例设计模式-1
class MP(object):
    instance = None       # 用类属性记录对象引用
    def __new__(cls,*args,**kwargs):
        if cls.instance is None:
            cls.instance = super().__new__(cls)
        return cls.instance
        
mp1 = MP()
mp2 = MP()
print(mp1)
print(mp2)
```

```python
# 实现单例设计模式-2
# __init__执行一次
class MP(object):
    instance = None 
    init_flag = False            # 用类属性记录对象引用
    def __new__(cls,*args,**kwargs):
        if cls.instance is None:
            cls.instance = super().__new__(cls)
        return cls.instance
    
    def __init__(self):
        if MP.init_flag:
            return
        print("初始化执行")
        MP.init_flag = True
        
        
mp1 = MP()
mp2 = MP()
print(mp1)
print(mp2)
```
