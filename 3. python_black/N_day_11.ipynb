# tcp-ip协议
- 协议族
- 协议为一种约定
- 连接五元组（来源IP，来源端口，目标IP，目标端口，协议）

- 应用层：HTTP，飞球...
- 传输层：TCP，UDP
- 网络层：（网际层）IP（网际层<----->应用层---原始套接字）
- 链路层：（网络接口层）网络接口


- 另一套标准：理论协议


# 路由器通信
- 路由器具有两个或两个以上网卡，不同网卡连接不同的交换机，具有不同交换机网段的ip
- 一个交换机下的数据发送到另一交换机下的设备时，目标IP及源IP不变，但目标mac及源mac会被路由器改变（即）
