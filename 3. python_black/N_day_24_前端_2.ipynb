## 列表
- `<ul><li></li></ul>`
- 列表样式list-style: none;

## Emment
- `ul>li*5`
- `ul>li{这是列表$}*5`
- `ul>(li>a[href='#']{这是链接})*5`
- `.box`-----创建一个class='box'的div标签
- `p.box`----创建一个class='box'的p标签
- `#box`-----创建一个id='box'的div标签
- `p#box`----创建一个id='box'的p标签

## id选择器`#`
- 一个标签只能有一个id,但可以有多个class(用空格隔开)
- 每个标签都可以设置id属性 id="id100"
- 链接id使用,href='#id100'
- 通过id设置样式,使用#id100{}

## 组/并集选择器`,`
- `.div1,.div2{}`写法可以同时修改div1和div2的样式

## 伪类及伪元素选择器(设置特效)
- 伪类操作标签------------------`:`
- 伪元素操作标签里的内容-----`::`
- div:hover{color:red}---------------鼠标悬浮时变红
- div::before{content:'这是文字'}--在元素前面加一段文字
- div::after{content:'这是文字'}----在元素后面加一段文字
- 伪元素添加的文字无法被选中

# CSS属性
- `text-align:center/right/left;`块元素中水平居中/右/左
- `text-indent:36px/2em;`首行缩进36px/2个字
- `font-style:italic;`文字斜体
- `font-weight:bold;`文字加粗
- `font-size:12px`文字大小
- `font-family:'Microsoft YaHei'`设置字体
- `font:italic bold 20px/30px 'Microsoft YaHei'`设置文字斜体,加粗,大小20px/行高30px,字体

# CSS元素溢出
- `overflow:visible/hidden/scroll/auto`默认值:内容不修剪/隐藏超出内容/使超出内容可滚动浏览/自动,滚动浏览(滚动条可自动显示和隐藏)

# CSS盒子模型
- width宽,height高,border边框,padding内边距,margin外边距
- 显示宽度=width+padding左右+border左右
- 显示高度=height+padding上下+border上下

## margin用法
- margin负值使边框合并
- margin:0 auto使其居中
-----------------
- 垂直外边距合并:margin-bottom:30px/margin-top:50px则两个块相距max(30,50)
- margin指的是边框对边框的距离,但是浮动后就是外边距对外边距
-----------------
- margin-top塌陷:子元素设定margin-top后,此属性转移给了父元素
- 原因:父元素因为没有设置边框导致子元素的外边距找不到参照
- 解决方法:1.设置边框 2.overflow 3.浮动 4.伪元素(推荐)
- 伪元素解决方法(父元素添加一个约定的class名字clearfix):.clearfix::before{content:'';display:table;}


-----------------
- 子元素水平垂直居中margin:计算 auto 0;
