Given a 32-bit signed integer, reverse digits of an integer.

Example 1:

Input: 123
Output:  321
Example 2:

Input: -123
Output: -321
Example 3:

Input: 120
Output: 21
Note:
Assume we are dealing with an environment which could only hold integers within the 32-bit signed integer range. For the purpose of this problem, assume that your function returns 0 when the reversed integer overflows.

```python
# 未使用算法，纯字符串操作
# 不足：需使用算法进行编程，还需研究
class Solution:
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        if x >= 2**31-1 or x <= -2**31:
            return 0
        elif x >= 0:
            x1 = str(x).replace(""," ")
            x2 = x1.strip(" ")
            x3 = x2.split(" ")
            x3.reverse()
            sum = ""
            for i in x3:
                sum += i
            if int(sum) >= 2**31-1:
                return 0
            else:
                return int(sum)
        else:
            x1 = str(-x).replace(""," ")
            x2 = x1.strip(" ")
            x3 = x2.split(" ")
            x3.reverse()
            sum = ""
            for i in x3:
                sum += i
            if int(sum) >= 2**31:
                return 0
            else:
                return -int(sum)
```
