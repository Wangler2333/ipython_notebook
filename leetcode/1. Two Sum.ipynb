Given an array of integers, return indices of the two numbers such that they add up to a specific target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

Example:
Given nums = [2, 7, 11, 15], target = 9,

Because nums[0] + nums[1] = 2 + 7 = 9,
return [0, 1].

```python
# 遍历列表，目标值减去元素的值和索引写入字典，每次循环都查找字典里是否有目标值减去元素的key
# 知识点：enumerate()枚举函数可以分别遍历索引及数据
class Solution:
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        dic = {}
        for index,items in enumerate(nums): # index为索引，items为数据
            if target-items in dic:
                return [dic[target-items],index] #查找字典里是否有差值
            else:
                dic[items] = index # 差值及索引加入字典
```
